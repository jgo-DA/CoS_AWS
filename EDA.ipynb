# EDA = Exploratory Data Analysis

```python
import pandas as pd, numpy as np
```

## Load the three source data files

```python
df_world = pd.read_csv('s3://flutz-bucket/spotify/world_data.csv', index_col=0)

display(df_world)
```

```python
df_brazil = pd.read_csv('s3://flutz-bucket/spotify/brazil_data.csv', index_col=0)

display(df_brazil)
```

```python
df_2020 = pd.read_csv('s3://flutz-bucket/spotify/2020_spotify.csv', index_col=0)

display(df_2020)
```

## Merge them into one dataframe - beware of duplicates!

```python
# Just concat all three tables together for a start

df_joined = pd.concat([df_world, df_brazil, df_2020]) \
  .sort_values(['popularity', 'track_id'], ascending=False)

display(df_joined)
```

```python
# Right, we quite certainly have complete duplicates in there - drop those (keeping one)

df_joined.drop_duplicates(inplace=True)

display(df_joined)
```

```python
# Knowing we sorted by popularity earlier, drop duplicates differing by popularity only (keep first=best row)

df_joined.drop_duplicates(subset=df_joined.columns.drop(['track_id', 'popularity']), inplace=True)

display(df_joined)
```

```python
# Now some more complicated duplicates - same artist, track_name and duration

subset = 'artist_name track_name duration_ms'.split()

df_joined[df_joined.duplicated(subset=subset, keep=False)].sort_values(subset)
```

```python
# Yeah, looks like it's safe to merge them too

df_joined.drop_duplicates(subset=subset, inplace=True)

display(df_joined)
```

```python
# Finally, replace row-id index by track_id from source data

df_joined.set_index('track_id', inplace=True)

display(df_joined)
```

```python
def my_info(dataFrame, showMinMax=False):
  dataFrame = dataFrame.copy().reset_index(); from pandas import option_context
  info = dataFrame.dtypes.to_frame('DType')
  info['Null'] = dataFrame.isna().sum().transform(lambda n:
    f'{n} ({n / len(dataFrame) * 100 :6.1f} %)'.replace(' ', '\xA0'*2))
  for idx in info.index:
    if (type(dataFrame[idx].iat[0]) not in [list, set]): series_tmp = dataFrame[idx];
    else: series_tmp = dataFrame[idx].astype(str)
    info.at[idx, 'nUnique'] = str(series_tmp.nunique(dropna=False))
    try: uniques = series_tmp.unique(); uniques.sort()
    except: uniques = sorted(uniques, key=repr)
    info.at[idx, f'Uniques [{len(dataFrame)} rows total]'] = str(uniques).replace('\n','')
    info.at[idx, 'Top_10'] = str(series_tmp.value_counts(dropna=False).head(10).to_dict())
  if (showMinMax): info = info.join(dataFrame.agg('min max median mean var std' \
    .split()).rename(str.capitalize).T)
  with option_context('max_colwidth', 275): display(info.rename_axis('Column').reset_index())
  dtypes = str(sorted(info['DType'].value_counts(sort=False).to_string().split('\n')))
  while '  ' in dtypes: dtypes = dtypes.replace('  ', ' ')
  print('DTypes: ' + dtypes[2:-2].replace(' ','(').replace("',('",'), ') + ')')

my_info(df_joined, True)
```

```python
import matplotlib.pyplot as plt

def draw_hist_plots(df, df_name='', subset=None):
    if (type(df) == str): df_name = df; df = globals()[df]
    if (df_name != ''): df_name += ' '

    if (subset is None): subset = df.select_dtypes('number').columns
    elif (type(subset) == str): subset = [subset]

    print(f'Total rows of data in {df_name}dataframe: {len(df):,}')
    if (len(df) < 1): return

    for i, column in enumerate(subset):
      if (i % 3 == 0): axes = plt.subplots(ncols=3, figsize=(22, 2.5))[1]
      ax = axes.flat[i % 3]; bins = df[column].nunique()

      if (bins > 15) or (df[column].max() - df[column].min() > 30):
        df[column].plot.hist(ax=ax, bins=10, rwidth=0.95,
                             title=f'Histogram of {column}', label='_')
      else: pd.DataFrame(index=np.arange(df[column].min(), df[column].max()+1)) \
        .join(df[column].value_counts(sort=False).rename('_')).plot.bar(ax=ax,
          width=0.9, title=f'Distribution of {column}', rot=0, legend=False); \
        ax.set_ylabel('Frequency')

      for line in [(df[column].mean(), 'red', '--', 'mean'),
                   (df[column].median(), 'orange', '-', 'median')]:
        ax.axvline(line[0], color=line[1], linestyle=line[2])
        ax.plot([], [], '-', color=line[1], linestyle=line[2], label=f'{line[3]}={line[0]:0,.4f}')

      yticks = ax.get_yticks(); xticks = ax.get_xticks(); ax.grid(axis='y'); ax.legend()  
      if (yticks[-1] > 9999): ax.set_yticklabels(f'{_:,.0f}' for _ in yticks)
      if (xticks[-1] > 9999): ax.set_xticklabels(_ and f'{_//1000:,.0f}k' or '0' for _ in xticks)

      va_split = ax.get_ylim()[1] * 0.5
      for patch in ax.patches: y = patch._height; \
        ax.annotate(f'  {y:0,.0f} \n  {y/len(df):0.1%} ', [patch.xy[0]+patch._width/1.75, y],
          weight=('bold' if y > va_split else 'normal'), va=('top' if y > va_split else 'bottom'),
          ha='center', color=('red' if y == 0 else ('white' if y > va_split else 'black')), rotation=90)

    for i in range(i % 3 + 1, 3): axes.flat[i % 3].set(xticks=[], yticks=[]) # hide trailing subplots
    plt.show()

draw_hist_plots('df_joined')
```
