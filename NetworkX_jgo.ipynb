```python
# Import the necessary libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

import networkx as nx
```

```python
# Load the dataset.
data = pd.read_csv('joined.csv', sep=",", header=0, index_col=0, nrows=1000)
data.head()
```

```python
# WHAT edges are requied 
G = nx.read_edgelist('joined.csv', delimiter=',', nodetype=str)
```

```python
# get network data
print(nx.info(G))
```

```python
# visalize network

plt.figure(figsize=(15,15))
nx.draw_networkx(G, node_size=15, alpha=0.3, node_color='r', edge_color='k', width=0.5, with_labels=False)
plt.show()
```
