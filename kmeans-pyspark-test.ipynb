```
from pyspark.sql.functions import *
```

```
def load_dataframe(file_path, seperator):
  df = spark.read.csv(file_path, header=True, inferSchema=True, sep=seperator)
  return df
```

```
df = load_dataframe('/FileStore/tables/world_data.csv', ',')
```

```
df.display()
```

```
features_cols = ['popularity','danceability','energy','key','loudness','mode','speechiness','acousticness','instrumentalness','liveness','valence', 'tempo', 'duration_ms','time_signature']

for col in features_cols:
  df = df.withColumn(col,df[col].cast('float'))
```

```
df = df.na.drop()
```

```
# df = df.withColumn("instrumentalness", df["instrumentalness"].cast('float'))
# df_num = df.withColumn(column, df[column].cast("float").alias(column) for column in df.columns[4:])
```

```
from pyspark.ml.feature import VectorAssembler

vecAssembler = VectorAssembler(inputCols=features_cols, outputCol="features")
df_kmeans = vecAssembler.transform(df).select('track_id', 'features')
```

```
from pyspark.ml.feature import MinMaxScaler
scaler = MinMaxScaler(inputCol="features", outputCol="scaledFeatures")
scalerModel = scaler.fit(df_kmeans.select("features"))
df_scaled = scalerModel.transform(df_kmeans)
```

```
from pyspark.ml.clustering import KMeans
import numpy as np

cost = np.zeros(20)

for k in range(2,20):
    kmeans = KMeans().setK(k).setSeed(1).setFeaturesCol('scaledFeatures')
    model = kmeans.fit(df_scaled)
    cost[k] = model.summary.trainingCost 
```

```
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator

fig, ax = plt.subplots(1,1, figsize =(8,6))
ax.plot(range(2,20),cost[2:20])
ax.set_xlabel('k')
ax.set_ylabel('cost')
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
plt.show()
```

```
k = 8
kmeans = KMeans().setK(k).setSeed(1).setFeaturesCol('scaledFeatures')
model = kmeans.fit(df_scaled)
```

```
centers = model.clusterCenters()

print("Cluster Centers: ")
for center in centers:
    print(center)
```

```
transformed = model.transform(df_scaled).select('track_id', 'prediction')
rows = transformed.collect()
```

```
df_pred = sqlContext.createDataFrame(rows)
df_pred.show()
```

```
df_pred = df_pred.join(df, 'track_id')
df_pred.show()
```

```
pddf_pred = df_pred.toPandas().set_index('track_id')
pddf_pred.head()
```

```
from mpl_toolkits.mplot3d import Axes3D

threedee = plt.figure(figsize=(12,10)).gca(projection='3d')
threedee.scatter(pddf_pred.popularity, pddf_pred.danceability, pddf_pred.energy, c=pddf_pred.prediction)
threedee.set_xlabel('popularity')
threedee.set_ylabel('danceability')
threedee.set_zlabel('energy')
plt.show()
```
